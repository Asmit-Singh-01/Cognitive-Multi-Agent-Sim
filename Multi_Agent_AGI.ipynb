```python
import numpy as np

class SurvivalEnv:
    def __init__(self, grid_size=5):
        # Duniya ka size (5x5 grid)
        self.grid_size = grid_size
        self.reset()

    def reset(self):
        # Game restart hone par agents aur resources ki position
        self.agent1_pos = [0, 0]           # Agent 1 top-left me
        self.agent2_pos = [self.grid_size-1, self.grid_size-1] # Agent 2 bottom-right me

        # Ek resource (khana/energy) random jagah par
        self.resource_pos = [np.random.randint(0, self.grid_size),
                               np.random.randint(0, self.grid_size)]

        return self._get_state()

    def _get_state(self):
        # AI ko batana ki wo kahan hai aur resource kahan hai (State estimation)
        state = [
            self.agent1_pos[0], self.agent1_pos[1],
            self.agent2_pos[0], self.agent2_pos[1],
            self.resource_pos[0], self.resource_pos[1]
        ]
        return np.array(state)

    def step(self, action1, action2):
        # Actions: 0=Up, 1=Down, 2=Left, 3=Right
        self.agent1_pos = self._move(self.agent1_pos, action1)
        self.agent2_pos = self._move(self.agent2_pos, action2)

        # Rewards (AI ko sikhane ke liye)
        reward1, reward2 = -0.1, -0.1 # Har step par thodi energy loss (survival pressure)
        done = False

        # Agar Agent 1 ne resource collect kiya
        if self.agent1_pos == self.resource_pos:
            reward1 += 10
            done = True

        # Agar Agent 2 ne resource collect kiya
        elif self.agent2_pos == self.resource_pos:
            reward2 += 10
            done = True

        return self._get_state(), (reward1, reward2), done

    def _move(self, pos, action):
        new_pos = list(pos)
        if action == 0 and pos[0] > 0: new_pos[0] -= 1                  # Up
        if action == 1 and pos[0] < self.grid_size - 1: new_pos[0] += 1 # Down
        if action == 2 and pos[1] > 0: new_pos[1] -= 1                  # Left
        if action == 3 and pos[1] < self.grid_size - 1: new_pos[1] += 1 # Right
        return new_pos

# Code check karne ke liye:
env = SurvivalEnv()
print("Initial World State:", env.reset())
print("Environment successfully created! Ready for AI Agents.")
```
